```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


%matplotlib inline
```

```python
df_avo = pd.read_csv("avoN.csv")
```

```python
df_avo.head(15)
```

```python
df_avo.info()
```

```python
x = df_avo['AveragePrice'].values
y = df_avo['Total Bags'].values

print("Precios Promedio máximo: ", df_avo['AveragePrice'].max())
print("Precios Promedio mínimo: ", df_avo['AveragePrice'].min())
print("Promedio de precios promedio: ", df_avo['AveragePrice'].mean())
```

```python
X = np.array(list(zip(x,y)))
print(X)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_ 
```

```python
colors = ["m.", "r.", "c.", "y.", "b."]
for i in range(len(X)):
    print("Coordenada: ", X[i], " Label: ", labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
plt.scatter(centroids[:,0],centroids[:,1], marker = 'x', s = 150, linewidths = 5, zorder = 10)
plt.show()
```
